# Проект по машинному обучению

**Цель исследования:**

Сделать прогноз, превышает ли годовой доход человека 50 000 долларов США в год, на основе данных переписи населения.
Датасет также известен как «Доход по переписи».


**Описание данных:**

Файлы `adult.data` и `adult.test` содержат данные о особенностях человека и его годовой доход.

- `age` - возраст человека.
- `workclass` - какой вид занятости у человека.

- `occupation` - род деятельности человека.
- `relationship` - отношение человека к главе домохозяйства.
- `race` - раса человека.
- `sex` - пол человека.
- `capital-gain` - доход от капитала за год.
- `capital-loss` - убытки от капитала за год.
- `hours-per-week` - количество рабочих часов в неделю.
- `native-country` - страна рождения.
- `income` - уровень дохода

### 2.1. Переименование названий столбцов

Напишем функцию для переименования столбцов к snake_case.

```python
def rename_column(df, names_column):
    dict_of_names = {}
    for name in names_column:
        if "-" in name:
            dict_of_names[name] = name.replace("-", "_")
    return df.rename(columns=dict_of_names)
```

Построим функцию, которая будет строить гистрограмму распределения и диаграмму размаха для нужно признака.

```python
def plot_hist_and_boxplot(df, x, bins=None, ylabel_hist='', title=''):
    if not bins:
        bins = int(np.log2(len(df))) + 1
    title = f"Гистрограмма и диаграмма размаха для признака - '{x}'"
    fig, ax = plt.subplots(1, 2)
    sns.histplot(data=df, x=x, bins=bins, edgecolor="black", ax=ax[0])
    sns.boxplot(data=df, x=x, orient='h', ax=ax[1], flierprops=dict(markerfacecolor="red"), notch=True)
    ax[0].set_xlabel(x)
    ax[1].set_xlabel(x)
    ax[0].set_ylabel(ylabel_hist)
    fig.suptitle(title)
    plt.show()
```

Также определим функцию для поиска количества выбросов для заданного признака и их процентного соотношения по отношению к общему объему выборки.

```python
def find_outliers(df, feature):
    q_1, q_3 = df[feature].quantile([0.25, 0.75])
    iqr = q_3 - q_1
    low_boundary = q_1 - iqr * 1.5
    upper_boundary = q_3 + iqr * 1.5

    cnt_fliers = len(df[(df[feature] > upper_boundary) | (df[feature] < low_boundary)])
    print(f"Количество выбросов - {cnt_fliers}")
    print(f"Процентное содержание выбросов - {cnt_fliers/len(df) * 100:.2f} %")
    
```

### 3.2. Категориальные признаки

Определим функцию, которая будет подсчитывать количество каждой категории для категориального признака.

```python
def plot_categorical(df, feature, hue=None):
    sns.countplot(df, y=feature, stat="percent", hue=hue)
    plt.show()
```

Определим функцию, которое будет возвращать процентное соотношение категорий в категориальном признаке.

```python
def eval_percent_for_category(df, feature, hue=None, value_for_percent=None):
    if not hue:
        return df[feature].value_counts() / len(df) * 100
    piv_df = df.pivot_table(index=feature, columns=hue, aggfunc="size", fill_value=0)
    piv_df["percent_" + value_for_percent] = piv_df[value_for_percent] / piv_df.sum(axis=1) * 100
    return piv_df
```

#### 3.2.1. Признак `workclass`

```python
plot_categorical(data_train, "workclass")
```

Посмотрим более подробно на процентное соотношение `workclass`:

```python
eval_percent_for_category(data_train, "workclass")
```

Большая часть людей имеют рабочий класс 'Private' - около 70 %.  Рабочий класс `Without-pay` и `Never-worked` имеет маленькое количество человек - <0.1 процента.

Посмотрим как распредлен годовой доход для каждого из рабочего класса.

```python
plot_categorical(data_train, "workclass", hue="income")
```

Посмотрим более подробно на процентное соотношение:

```python
eval_percent_for_category(data_train, "workclass", hue="income", value_for_percent=">50K")
```

```python
def plot_scatters(df, hue):
    features = df.select_dtypes(exclude="object").columns
    features = features.drop(hue)
    features_tuples = []
    for feature_1 in features:
        for feature_2 in features:
            if feature_1 != feature_2 and (feature_2, feature_1) not in features_tuples:
                features_tuples.append((feature_1, feature_2))
    n_graphs = len(features_tuples)
    fig, axes = plt.subplots(n_graphs // 2, 2, figsize=(20, 20))
    i = 0
    j = 0
    for k, (feature_1, feature_2) in enumerate(features_tuples):
        sns.scatterplot(data=df, x=feature_1, y=feature_2, hue=hue, ax=axes[j, i])
        j += 1
        if j > n_graphs // 2 - 1:
            j = 0
            i += 1
    fig.suptitle("Диаграммы рассеяния для количественных признаков")
        
```

Построим диаграмму рассеяния для количественных признаков.

```python
def plot_scatter_for_target(df, feature, target):
    sns.scatterplot(data=df, x=feature, y=target)
    plt.title(f"Диаграмма рассеяния для {target} и {feature}")
```

### 5.2. Инициализация преобразователей данных

Разделим обучающие данные на матрицу признаков и вектор таргетов.

```python
X = data_train.drop(columns="income")
y = (data_train["income"]).astype(int)
X.head()
```

Построим ColumnTransformer, где стандартизуем числовые признаки и проведем One-Hot Encoding над категориальными.

```python
numeric_col = X.select_dtypes(include="number").columns
print("Наименования столбцов с количественными признаками:\n", numeric_col, end="\n\n")
categorical_col = X.select_dtypes(exclude="number").columns
print("Наименования столбцов с категориальными признаками:\n", categorical_col)
```

```python
col_transformer = ColumnTransformer([('scaler', StandardScaler(), numeric_col),
                                     ("ohe", OneHotEncoder(handle_unknown="ignore", sparse_output=False), categorical_col)],
                                    remainder="passthrough")
```

Посмотрим как будет выглядеть набор признаков после применения преобразований.

```python
pd.DataFrame(col_transformer.fit_transform(X), columns = col_transformer.get_feature_names_out())
```

Заметим, что мы не указали параметр `drop="fisrt"` в OneHotEncoder. Для 'деревянных' моделей это не вызовет сложностей, однако логистическая регрессия будет давать большие коэффициенты, так как признаки мультиколлинеарны, поэтому будем бороться с мультиколлинеарностью регуляризацией.
Если бы мы указали этот параметр, то категории, которые модель еще не видела кодировались бы нулями, а также некоторые из объектов, которые модель уже видела из-за `drop="fisrt"` кодировалась бы нулями, но это не одно и тоже, поэтому этот параметр решено было убрать.

А давайте сразу проверим, встречаются ли категории в тестовом датасете, которых не было в обучающем, и верно ли они кодируются. Для этого пройдемся по всем категориальным столбцам и посмотрим совпадают ли уникальные значения.

### 5.3. Создание и обучение pipeline

В качестве целевой метрики будем использовать `precision`. Постараемся добиться максимальной precision при `recal >= 0.6`.

Опишем функцию, которая будет собирать конвеер машинного обучения для модели и преобразователя данных, искать наилучшие гиперпараметры при кросс-валидации. Функция будем возвращать обученный pipeline.

```python
def create_and_evaluate_pipeline(model, model_alias, col_transformer, params_for_grid, search_method="GridSearch", report=True):
    if search_method not in ["GridSearch", "RandomizedSearch"]:
        raise ValueError(f"Неверное значение для параметра search_method. Возможные значения: {['GridSearch', 'RandomizedSearch']}")
    col_transformer_alias = "col_transfomer"
    pipeline = Pipeline([(col_transformer_alias, col_transformer),
                         (model_alias, model)])

    if isinstance(params_for_grid, dict):
        params = {}
        for key in params_for_grid:
            params[model_alias + "__" + key] = params_for_grid[key]
    else:
        params = []
        for lst in params_for_grid:
            cur_params = {}
            for key in lst:
                cur_params[model_alias + "__" + key] = lst[key]
            params.append(cur_params)

    if search_method == "GridSearch":
        cv_model = GridSearchCV(estimator=pipeline,
                         param_grid=params,
                         scoring="average_precision",
                         cv=5,
                         refit=True,
                         n_jobs=-1,
                         verbose=2)
    else:
        cv_model = RandomizedSearchCV(estimator=pipeline,
                               param_distributions=params,
                               scoring="average_precision",
                               cv=5,
                               refit=True,
                               n_jobs=-1,
                               verbose=2,
                               random_state=random_state,
                               n_iter=200)
    t_start = time.time()
    cv_model.fit(X, y)
    t_end = time.time()
    t_dur = t_end - t_start
    if report:
        print(f"Лучшие параметры - {cv_model.best_params_}\n")
        print(f"Лучшая метрика при кросс-валидации - {cv_model.best_score_:.3f}\n")
        print(f"Время, потраченное на подбор оптимальных гиперпараметров - {t_dur // 60} мин. {t_dur % 60:.2f} сек.")
    return cv_model
        
```

### 5.4. Сравнение моделей машинного обучения

### 5.4.1. Сравнение моделей по непороговым метрикам

Для каждой модели посчитаем ROC AUC и Average precision и сравним модели между собой.

Для этого опишем функцию, которая будет принимать список моделей и вернет датафрейм, где индексами будут названия моделей, а столбцами метрики.

```python
def calc_nonthreshold_metrics(models, models_names, X_test, y_test):
    col_names = ["roc_auc", "avg_prec"]
    res_df = pd.DataFrame(index=models_names, columns=col_names, dtype="float64")
    for ind, model in enumerate(models):
        model_predict = model.predict_proba(X_test)[:, 1]
        res_df.loc[models_names[ind]] = [roc_auc_score(y_test, model_predict),
                                         average_precision_score(y_test, model_predict)]
    return res_df
```

Разделим тестовые данные на матрицу признаков и вектор таргетов.

```python
X_test = data_test.drop(columns="income")
y_test = (data_test["income"] == '>50K.').astype(int)
X_test.head()
```

```python
y_test.head()
```

```python
pipelines = [knn_pipeline, lr_pipeline, dt_pipeline, rf_pipeline, xgb_pipeline]
pipe_names = ["knn", "log_regression", "decision_tree", "random_forest", "xgboost"]
nonthres_metrics = calc_nonthreshold_metrics(pipelines,
                                             pipe_names,
                                             X_test,
                                             y_test)
```

Посмотрим на непороговые метрики для наших моделей.

```python
nonthres_metrics
```

#### 5.4.2. Сравнение моделей по пороговым метрикам.

Построим вспомогательную функцию, которая строит датафрейм, содержащий пороги и пороговые метрики. Также она может вернуть и оптимальный порог с соответствующими ему метриками. Здесь учитываеся условие `recall >=0.6`.

```python
def find_threshold(model, X_test, y_test, recall_threshold, return_optim=False):
    col_names = ["precision", "recall", "f1", "accuracy", "balanced_accuracy"]
    thresholds = np.linspace(1e-5, 1, 500)
    res_df = pd.DataFrame(index=thresholds, columns=col_names, dtype="float64")
    predict_proba = model.predict_proba(X_test)[:, 1]
    for threshold in thresholds:
        predict = (predict_proba >= threshold).astype(int)
        precision = precision_score(y_test, predict)
        recall = recall_score(y_test, predict)
        f1 = f1_score(y_test, predict)
        accuracy = accuracy_score(y_test, predict)
        balanced_accuracy = balanced_accuracy_score(y_test, predict)

        res_df.loc[threshold] = [precision, recall, f1, accuracy, balanced_accuracy]
    res_df = res_df[res_df["recall"] >= recall_threshold].sort_values(by="precision", ascending=False).reset_index(names="threshold")
    if not return_optim:
        return res_df
    return res_df.loc[0]
    
        
```

Опишем функцию, которая будет искать оптимальный порог при условии, что `recall` >= 0.6, а `precision` мы хотим максимизировать. Она вернет датафрейм, содержащий пороговые метрики для каждой модели, а также выбранный порог.

```python
def calc_threshold_metrics(models, models_names, X_test, y_test, recall_threshold):
    col_names = col_names = ["threshold", "precision", "recall", "f1", "accuracy", "balanced_accuracy"]
    res_df = pd.DataFrame(index=models_names, columns=col_names, dtype="float64")
    for ind, model in enumerate(models):
        res_df.loc[models_names[ind]] = find_threshold(model, X_test, y_test, recall_threshold, return_optim=True)
    return res_df.sort_values(by="precision", ascending=False)
```

Посмотрим на возможные варианты порогов для модели XGBoost при нашем условии на `полноту`.

```python
find_threshold(xgb_pipeline, X_test, y_test, recall_threshold=0.6)
```

Теперь сравним модели по их пороговым метрикам на оптимальном пороге.

```python
pipelines = [knn_pipeline, lr_pipeline, dt_pipeline, rf_pipeline, xgb_pipeline]
pipe_names = ["knn", "log_regression", "decision_tree", "random_forest", "xgboost"]
thres_metrics = calc_threshold_metrics(pipelines, pipe_names, X_test, y_test, 0.6)
```

```python
thres_metrics
```

1. Самая лучшая модель `xgboost` показала `precision` = 0.82.
2. Следующая по качеству модель - `случайный лес` (`precision` = 0.82).
3. Как и ожидалось, ансамблевые модели показали самый хороший результат.
